```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from cmapPy.pandasGEXpress.parse import parse
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = parse("PAAD.gct")
df = data.data_df
row_metadata = data.row_metadata_df
col_metadata = data.col_metadata_df
print("Content of given GCT file")
df 
```

```python
total_genes = len(df.index)
print("Total Genes : ", total_genes)
```

```python
# drop all genes with nan values
df = df.dropna()
print("Cleaned Genes Data : ")
df
```

```python
print("Number of valid genes with no Nan values : ",len(df.index))
print("Number of genes with Nan values : ", total_genes - len(df.index))
```

```python
print("Generating gene expression distribution for all samples ")
for (columnName, columnData) in df.iteritems():
    sns.distplot(columnData, hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = columnName) 
plt.legend(prop={'size': 5}, title = 'Samples')
# plt.title('Density Plot with Multiple Airlines')
plt.xlabel('Sample Data')
plt.ylabel('Density')
plt.title('Genes Distribution for all samples')
```

```python
print("Content of Row metadata in a given GCT file")
row_metadata
```

```python
print("Content of Column metadata in a given GCT file")
col_metadata
```

```python
# Transposing the matrix for easier concatenation of df to col_metadata matrix
df=df.T

# scale the features in the data before applying PCA
scalar=StandardScaler()
scalar.fit(df)
```

```python
# Standardizing the features
scaled_data=scalar.transform(df)
```

```python
print("Standardised Scaled Data of genes for different sample : ")
scaled_data
```

```python
# initialising pca with 2 components
pca=PCA(n_components=2)
```

```python
# fitting the scaled data in pca
pca.fit(scaled_data)
```

```python
# applying pca transform
x_pca=pca.transform(scaled_data)
```

```python
print("Transformed matrix with two principal components of the data : ")
x_pca
```

```python
# generating dataframe 
principal_Df = pd.DataFrame(data = x_pca
             , columns = ['principal component 1', 'principal component 2'])
```

```python
print("DataFrame of the result PCA : ")
principal_Df
```
